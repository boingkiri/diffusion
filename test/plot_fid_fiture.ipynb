```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python


def plot_fid_scores(fid_files_dict):
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(1, 1, 1)
    
    fid_key = "Torso FID score"
    for legend, file_path in fid_files_dict.items():
        fid_data = pd.read_csv(file_path)
        corrupted_fid_key = ""
        for col in fid_data.columns:
            if fid_key in col:
                corrupted_fid_key = col
                break
        fid_data = fid_data.rename(columns={corrupted_fid_key: fid_key})
        fid_data.dropna()
        plt.plot(fid_data['Step'], fid_data['Torso FID score'], label=legend)

    plt.xlabel('Training Step (x10000)', fontsize=15)
    plt.ylabel('FID', fontsize=15)
    plt.yscale("log")
    plt.ylim(4, 200)
    plt.legend(fontsize=15)

    ax.set_xticks([0, 100000, 200000, 300000, 400000])
    ax.set_xticklabels(['0', '10', '20', '30', "40"], fontsize=12)
    ax.set_yticks([4, 10, 20, 40, 100])
    ax.set_yticklabels(['4', '10', '20', '40', "100"], fontsize=12)

    plt.margins(x=0.01)

    # ax.set_yticks([0, 10, 20, 40, 80])
    # ax.set_yticklabels(['0', '10', '20', '40', "80"], fontsize=12)
    
    # plt.title('FID Scores Over Training Steps')
    plt.grid(True)
    plt.show()
```

```python
fid_files = {
    ' 1': 'lambda/lamb1_c_003.csv',
    ' 0.3': 'lambda/lamb03_c_003.csv',
    ' 0.1': 'lambda/lambda_01_c_003.csv',
    ' 0.05': 'lambda/lamb005_c_003.csv',
}

plot_fid_scores(fid_files)
```

```python
# Example usage:
fid_files = {
    'c 0.03': 'lambda/lamb03_c_003.csv',
    'c 0.01': 'c/0.3_0.01.csv',
    'c 0.003': 'c/0.3_0.003.csv',
    'c 0.0003': 'c/0.3_0.0003.csv',
}

plot_fid_scores(fid_files)
```
